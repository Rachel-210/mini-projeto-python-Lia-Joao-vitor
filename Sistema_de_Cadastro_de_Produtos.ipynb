```python
estoque = []
codigos_cadastrados = set()
categorias = ("Alimentos", "Limpeza", "Bebidas", "Brinquedos")
proximo_codigo = 1

def validar_inteiro(prompt):
    while True:
        entrada = input(prompt).strip()
        if entrada.isdigit():
            valor = int(entrada)
            if valor >= 0:
                return valor
            else:
                print("O valor não pode ser negativo, digite um número inteiro positivo.")
        else:
            print("Error. Por favor, digite apenas números inteiros (0 ou positivo).")

def validar_float(prompt):
    while True:
        entrada = input(prompt).strip().replace(',', '.')
        if entrada:
            return float(entrada)
        else:
            print("Por favor, digite um valor.")

def validar_categoria():
    print("Categorias disponíveis:")
    for i, categoria in enumerate(categorias):
        print(f"{i+1} - {categoria}")

    num_categorias = len(categorias)
    while True:
        escolha = input(f"Escolha a categoria (1 a {num_categorias}): ").strip()
        if escolha.isdigit():
            indice = int(escolha) - 1
            if 0 <= indice < num_categorias:
                return categorias[indice]
            else:
                print("Escolha não está dentro do intervalo. Tente novamente.")
        else:
            print("Error. Por favor, digite o número da categoria.")

def cadastrar_produto_simples():
    global proximo_codigo
    print("Cadastro de Produto")

    codigo = proximo_codigo

    nome = input("Nome: ").strip()
    if not nome:
        print("Error. Por favor, digite um nome.")
        return

    preco = validar_float("Preço: ")
    quantidade = validar_inteiro("Quantidade: ")
    categoria = validar_categoria()

    novo_produto = {
        "codigo": codigo,
        "nome": nome,
        "preco": preco,
        "quantidade": quantidade,
        "categoria": categoria
    }

    estoque.append(novo_produto)
    codigos_cadastrados.add(codigo)
    proximo_codigo += 1

    print(f"Produto '{nome}' cadastrado com sucesso! Código: {codigo}")

def listar_produtos_simples():
    print("Lista de Produtos")
    if not estoque:
        print("A lista de produtos está vazia.")
        return

    print(f"{'Código':<8} {'Nome':<25} {'Preço':<10} {'Qtd.':<6} {'Categoria':<15}")

    for produto in estoque:
        preco_formatado = f"R$ {produto['preco']:.2f}"
        print(f"{produto['codigo']:<8} {produto['nome']:<25} {preco_formatado:<10} {produto['quantidade']:<6} {produto['categoria']:<15}")


def buscar_produto_simples():
    print("Buscar Produto")
    if not estoque:
        print("Estoque vazio.")
        return

    termo = input("Digite o código ou nome para buscar: ").strip()
    if not termo:
        print("Error. Por favor, você deve digitar o código ou o nome.")
        return

    resultados = []

    if termo.isdigit():
        codigo_busca = int(termo)
        for produto in estoque:
            if produto["codigo"] == codigo_busca:
                resultados.append(produto)
                break
    else:
        termo_lower = termo.lower()
        for produto in estoque:
            if termo_lower in produto["nome"].lower():
                resultados.append(produto)

    if resultados:
        print(f"{len(resultados)} Produto(s) encontrado(s):")

        print(f"{'Código':<8} {'Nome':<25} {'Preço':<10} {'Qtd.':<6} {'Categoria':<15}")

        for produto in resultados:
            preco_formatado = f"R$ {produto['preco']:.2f}"
            print(f"{produto['codigo']:<8} {produto['nome']:<25} {preco_formatado:<10} {produto['quantidade']:<6} {produto['categoria']:<15}")
    else:
        print(f"Nenhum produto encontrado para '{termo}'")

def atualizar_produto_simples():
    print("Atualizar Produto")

    while True:
        codigo_str = input("Digite o código do produto para atualizar: ").strip()
        if codigo_str.isdigit():
            codigo_busca = int(codigo_str)
            break
        print("Código inválido. Deve ser um número inteiro.")

    produto_atualizar = None
    for produto in estoque:
        if produto["codigo"] == codigo_busca:
            produto_atualizar = produto
            break

    if produto_atualizar:
        print(f"Atualizando: {produto_atualizar['nome']}")
        print("Deixe em branco para manter o valor atual.")

        novo_preco_str = input(f"Novo preço (valor atual: R$ {produto_atualizar['preco']:.2f}): ").strip().replace(',', '.')
        if novo_preco_str:
             produto_atualizar['preco'] = float(novo_preco_str)

        while True:
            nova_quantidade_str = input(f"Nova quantidade (valor atual: {produto_atualizar['quantidade']}): ").strip()
            if not nova_quantidade_str:
                break
            if nova_quantidade_str.isdigit():
                nova_quantidade = int(nova_quantidade_str)
                if nova_quantidade >= 0:
                    produto_atualizar['quantidade'] = nova_quantidade
                    break
                else:
                    print("A quantidade deve ser um número inteiro positivo.")
            else:
                 print("Error. Por favor, digite apenas números inteiros.")

        mudar_cat = input("Deseja mudar a categoria? (Digite sim ou nao): ").strip().lower()
        if mudar_cat == 'sim':
            nova_categoria = validar_categoria()
            produto_atualizar['categoria'] = nova_categoria

        print(f"Produto {codigo_busca} atualizado!")
    else:
        print(f"Produto com código {codigo_busca} não encontrado.")

def excluir_produto_simples():
    print("Excluir Produto")

    while True:
        codigo_str = input("Digite o código do produto para excluir: ").strip()
        if codigo_str.isdigit():
            codigo_busca = int(codigo_str)
            break
        print("Error. Po favor, o número deve ser inteiro.")

    indice_para_remover = -1
    for i, produto in enumerate(estoque):
        if produto["codigo"] == codigo_busca:
            indice_para_remover = i
            break

    if indice_para_remover != -1:
        produto_removido = estoque.pop(indice_para_remover)
        codigos_cadastrados.remove(codigo_busca)
        print(f"Produto '{produto_removido['nome']}' (Código: {codigo_busca}) excluído!")
    else:
        print(f"Produto com código {codigo_busca} não encontrado.")

while True:
    print("Sistema de Controle de Estoque")

    print("1 - Cadastrar produto")
    print("2 - Listar produtos")
    print("3 - Buscar produto")
    print("4 - Atualizar produto")
    print("5 - Excluir produto")
    print("0 - Sair")


    escolha = input("Escolha uma opção: ").strip()

    if escolha == '1':
        cadastrar_produto_simples()
    elif escolha == '2':
        listar_produtos_simples()
    elif escolha == '3':
        buscar_produto_simples()
    elif escolha == '4':
        atualizar_produto_simples()
    elif escolha == '5':
        excluir_produto_simples()
    elif escolha == '0':
        print("Você saiu!")
        break
    else:
        print("Opção inválida. Escolha de 0 a 5.")
```
