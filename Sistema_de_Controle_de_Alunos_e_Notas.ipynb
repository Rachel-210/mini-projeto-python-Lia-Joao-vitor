```python

alunos = {}
nomes_cadastrados = set()  # Conjunto para evitar duplicações de nome


# Função para cadastrar um novo aluno (sem notas)
def cadastrar_aluno():
    nome = input("Nome do aluno: ").title()
    if nome in nomes_cadastrados:
        print(f"O aluno {nome} já está cadastrado! Tente outro nome.")
        return

    nomes_cadastrados.add(nome)
    alunos[nome] = {"notas": [], "media": 0, "status_aluno": "Sem notas"}
    print(f"\nAluno {nome} cadastrado com sucesso!\n")


# Função para registrar notas
def registrar_notas():
    nome = input("Digite o nome do aluno: ").title()
    if nome not in alunos:
        print("\nAluno não encontrado! Cadastre o aluno primeiro.\n")
        return

    qtd_notas = int(input("Quantas notas deseja cadastrar? "))
    notas = []

    for i in range(1, qtd_notas + 1):
        nota = float(input(f"Digite a {i}ª nota (use vírgula ou ponto): ").replace(",", "."))
        notas.append(nota)

    media = sum(notas) / len(notas)
    status_aluno = "Aprovado!" if media >= 7 else "Reprovado!"

    alunos[nome]["notas"] = notas
    alunos[nome]["media"] = media
    alunos[nome]["status_aluno"] = status_aluno

    print(f"\nNotas cadastradas para {nome} com sucesso!")
    print(f"Média: {media:.2f} - {status_aluno}\n")


# Função para listar alunos e médias
def listar_alunos_medias():
    if not alunos:
        print("\nNenhum aluno cadastrado ainda.\n")
        voltar_menu()
        return

    print("\n--- Lista de Alunos e Médias ---")
    for nome, dados in alunos.items():
        print(f"{nome} - Notas: {dados['notas']} - Média: {dados['media']:.2f} - Status: {dados['status_aluno']}")
    voltar_menu()


# Função para buscar um aluno específico
def buscar_aluno():
    nome = input("Digite o nome do aluno: ").title()
    if nome in alunos:
        dados = alunos[nome]
        print(f"\nAluno: {nome}")
        print(f"Notas: {dados['notas']}")
        print(f"Média: {dados['media']:.2f}")
        print(f"Status: {dados['status_aluno']}\n")
    else:
        print("\nAluno não encontrado!\n")
    voltar_menu()


# Função para mostrar apenas aprovados e reprovados
def mostrar_aprovados_reprovados():
    if not alunos:
        print("\nNenhum aluno cadastrado ainda.\n")
        voltar_menu()
        return

    print("\n--- Alunos Aprovados e Reprovados ---")
    for nome, dados in alunos.items():
        print(f"{nome}: {dados['status_aluno']} (Média: {dados['media']:.2f})")
    voltar_menu()


# Função de relatórios gerais
def relatorios():
    if not alunos:
        print("\nNenhum aluno foi encontrado!\n")
        voltar_menu()
        return

    total_alunos = len(alunos)
    soma_medias = 0
    aprovados = 0
    reprovados = 0

    for dados in alunos.values():
        soma_medias += dados["media"]
        if dados["status_aluno"] == "Aprovado!":
            aprovados += 1
        elif dados["status_aluno"] == "Reprovado!":
            reprovados += 1

    media_geral = soma_medias / total_alunos
    print("\n--- RELATÓRIO GERAL ---")
    print(f"Total de alunos: {total_alunos}")
    print(f"Aprovados: {aprovados}")
    print(f"Reprovados: {reprovados}")
    print(f"Média geral da turma: {media_geral:.2f}\n")
    voltar_menu()


# Função para voltar ao menu principal
def voltar_menu():
    while True:
        resposta = input("Deseja voltar ao menu principal? (S/N): ").lower()
        if resposta == "s":
            break
        elif resposta == "n":
            print("\nEncerrando o sistema... Até mais!")
            exit()
        else:
            print("Opção inválida! Digite apenas 's' ou 'n'.")


# Menu principal atualizado
def menu():
    while True:
        print("\n---- Sistema de Controle de Alunos ----")
        print("1 - Cadastrar aluno")
        print("2 - Registrar notas")
        print("3 - Listar alunos e médias")
        print("4 - Buscar aluno")
        print("5 - Mostrar aprovados e reprovados")
        print("6 - Relatórios")
        print("0 - Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            cadastrar_aluno()
        elif opcao == "2":
            registrar_notas()
        elif opcao == "3":
            listar_alunos_medias()
        elif opcao == "4":
            buscar_aluno()
        elif opcao == "5":
            mostrar_aprovados_reprovados()
        elif opcao == "6":
            relatorios()
        elif opcao == "0":
            print("\nSaindo... Até mais!")
            break
        else:
            print("\nOpção inválida! Tente novamente.\n")


# Inicia o sistema
menu()
```
