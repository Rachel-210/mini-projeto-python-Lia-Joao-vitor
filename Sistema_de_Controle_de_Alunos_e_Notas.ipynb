```python
alunos = {}

# Função para cadastrar um novo aluno

def cadastrar_aluno():
    nome = input("Nome do aluno: ").title()
    qtd_notas = int(input("Quantas notas deseja cadastrar? "))

    if nome in alunos:
        print(f"O aluno {nome} já está cadastrado! Tente outo nome.")
        return

# Coleta as notas digitadas

    notas = []

    for i in range(1, qtd_notas + 1):
        nota = float(input(f"Digite a {i}ª nota: "))
        notas.append(nota)

# Calcula a média das notas

    media = sum(notas) / len(notas)

    if media >= 7:
        status_aluno = "Aprovado!"
    else:
        status_aluno = "Reprovado!"

# Armazena as informações do aluno dentro do dicionário principal

    alunos[nome] = {"notas": notas, "media": media, "status_aluno": status_aluno}
    print(f"\nAluno {nome} cadastrado com sucesso! Média: {media:.2f}\n")

# Função para listar todos os alunos

def listar_alunos():
    if not alunos:
        print("\nNenhum aluno cadastrado ainda.\n")
        voltar_menu()
        return

    print("\n--- Lista de Alunos ---")

    for nome, dados in alunos.items():
        print(f"\n{nome} - Notas: {dados['notas']} - Média: {dados['media']:.2f} - Status do aluno: {dados['status_aluno']}\n")

# Pergunta se o usuário quer voltar ao menu

    voltar_menu()

# Função para consultar um aluno específico

def consultar_aluno():
    nome = input("Digite o nome do aluno: ").title()

# Verifica se o aluno existe no dicionário

    if nome in alunos:
        dados = alunos[nome]
        print(f"\nAluno: {nome}")
        print(f"Notas: {dados['notas']}")
        print(f"Média: {dados['media']:.2f}\n")
        print(f"Status do aluno: {dados['status_aluno']}")
    else:
        print("\nAluno não encontrado!\n")
    voltar_menu()

# Função para gerar um relatório geral

def relatorio():

# Verifica se há alunos cadastrados

    if not alunos:
      print("\nNenhum aluno foi encontrado!")
      return

# Variáveis para o relatório

    total_alunos = len (alunos)
    media_alunos = 0
    aprovados = 0
    reprovados = 0

# Percorre todos os alunos e soma as médias, e ontar aprovados e reprovados

    for dados in alunos.values():
      media_alunos += dados["media"]
      if dados["media"] >= 7:
        aprovados += 1
      else:
        reprovados += 1

 # Calcula a média geral

    media_geral = media_alunos / total_alunos

    print("\n--- RELATÓRIO ---")
    print(f"Total de alunos: {total_alunos}")
    print(f"Aprovados: {aprovados}")
    print(f"Reprovados: {reprovados}")
    print(f"Média geral da turma: {media_geral:.2f}")

# Pergunta se o usuário quer voltar ao menu

    voltar_menu()

#Função para voltar ao menu

def voltar_menu():
    while True:
        resposta = input("\nDeseja voltar ao menu principal? (S/N): ").lower()
        if resposta == "s":
            break
        elif resposta == "n":
            print("\nEncerrando o sistema... Até mais!")
            exit()
        else:
            print("\nOpção inválida! Digite apenas 's' ou 'n'.")

#Menu principal

def menu():
    while True:
        print("\n---- Sistema de Controle de Alunos ----")
        print("1 - Cadastrar aluno")
        print("2 - Listar todos os alunos")
        print("3 - Consultar aluno")
        print("4 - Relatório")
        print("5 - Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            cadastrar_aluno()
        elif opcao == "2":
            listar_alunos()
        elif opcao == "3":
            consultar_aluno()
        elif opcao == "4":
            relatorio()
        elif opcao == "5":
            print("\nSaindo... Até mais!")
            break
        else:
            print("\nOpção inválida! Tente novamente.\n")

# Chama o menu principal para iniciar o sistema

menu()
```
